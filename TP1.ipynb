## Descente de gradient à pas fixe

On rappelle que la méthode de gradient à pas fixe $\alpha$ pour une fonctionnelle $f: \mathbb{R^n} \longrightarrow \mathbb{R}$ consiste en le calcul des itérés :
$$x_{k+1} = x_k - \alpha \nabla f(x_k)$$
pour un $x_0$ donné.

Un critère d'arrêt classique est d'itérer jusqu'à ce que la norme $l_2$ du gradient soit inférieure à une tolérance donnée.
On limite également le nombre d'itérations total pour éviter les boucles infinies.


**Mettre en œuvre** la méthode du gradient à pas fixe pour la fonction de Rosenbrock suivante en dimension 2 : 
$$
f(x,y) = (1 - x)^2 + 100(y - x^2)^2
$$
en partant de $x_0 = (0, 2)$, avec 1000 itérations maximum, un pas de descente de $10^{-3}$ et une tolérance de $10^{-4}$.

**Sur une figure** de l'espace sur lequel $f$ est définie, en se restreignant au carré $[-1.5, 2.5]^2$, placer les itérés successifs de la méthode (ceci nécessite d'enregistrer les positions dans l'algorithme de descente) et le minimum global de la fonction de Rosenbrock (calculé formellement) afin d'illustrer la convergence de la méthode. On pourra également représenter des lignes de niveau de la fonction. 


```python
# Définition de la fonction f
def rosenbrock(x, y):
    return (1-x)**2 + 100 * (y-x**2)**2
```

```python
def grad_rosenbrock(x, y):
    grad = np.zeros(2)
    grad[0] = -2 * (1 - x) - 400 * x * (y - x**2)
    grad[1] = 200 * (y - x**2)
    return grad
```

```python
# Algorithme du gradient à pas fixe.
def gradient_fixe(grad, x0, rho, it_max, tol):
    x = x0
    for i in range(it_max):
        direction = grad(x[0], x[1])
        x = x - rho * grad(x[0], x[1])
        if npl.norm(direction) < tol:
            print("Arrivé en à convergence", i, "itérations")
            break
    return x

    
    
```

## Descente de gradient avec backtracking
Pour la même fonction de Rosenbrock, on veut utiliser la méthode du gradient avec *backtracking* pour satisfaire la condition d'Armijo :

$$f(x_k + t_k d_k) \le f(x_k) + t_k \omega \langle \nabla f(x_k), d_k \rangle$$
pour une direction $d_k$, un pas de descente $t_k$ et un paramètre $\omega \in (0, 1)$.

Voici une illustration de cette règle:

```python
def backtracking(f, gradf, x, rho, omega = 0.5, beta = 0.1) :
    gf = gradf(x[0], x[1])
    while not f(x[0] - rho * gf[0], x[1] - rho * gf[1]) <= f(x[0], x[1] ) - rho * omega * npl.norm(gf)**2:
        rho = rho * betat
    return rho    
    
```

```python


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
#%matplotlib notebook
%matplotlib inline
# ou "%matplotlib tk", par exemple avec VSCode

# Choix de la fonction
case = "cos"

if case == "cos":
    def f(x):
        return np.cos(2*x)

    def gradf(x):
        return -2*np.sin(2*x)

    x0 = 0.3
    xmin, xmax = 0, 4
    c = 0.1
    
elif case == "quad":
    def f(x):
        return x**2

    def gradf(x):
        return 2*x

    x0 = -1.8
    xmin, xmax = -2, 2
    c = 0.1

# Règle d'Armijo
def armijo(x0, c, x):
    return f(x0) + c * (x - x0) * gradf(x0)

# Discrétisation de l'espace des paramètres
x = np.linspace(xmin, xmax, 101)

# Figure initiale
plt.figure(figsize=(8, 8))
plt.plot(x, f(x), label="$f(x)$")
x0_graph, = plt.plot(x0, f(x0), 'o', markersize=10)
c_graph, = plt.plot(x, armijo(x0, c, x))
plt.title("Armijo rule")

plt.subplots_adjust(left=0.1, bottom=0.35)
c_axe = plt.axes([0.1, 0.1, 0.8, 0.05])
c_slider = Slider(
    ax=c_axe,
    label='c',
    valmin=-0.1,
    valmax=1.,
    valinit=c,
)

x0_axe = plt.axes([0.1, 0.17, 0.8, 0.05])
x0_slider = Slider(
    ax=x0_axe,
    label='$x_0$',
    valmin=xmin,
    valmax=xmax,
    valinit=x0,
)

def update(_):
    x0_graph.set_xdata(x0_slider.val)
    x0_graph.set_ydata(f(x0_slider.val))
    c_graph.set_ydata(armijo(x0_slider.val, c_slider.val, x))

c_slider.on_changed(update)
x0_slider.on_changed(update)
plt.show()
```

La méthode de **backtracking** consiste alors, **à chaque étape de gradient**, de trouver le pas de descente $t_k$ le plus grand qui satisfasse cette condition.
Une manière de faire est de multiplier un pas initial (trop grand à priori) par un facteur $\beta \in (0, 1)$ jusqu'à ce que le pas soit acceptable **pour l'itération courante**.


**Mettre en œuvre cette méthode et comparer**

```python
# Algorithme du gradient avec backtracking. 
```

```python
def gradient_back(f, gradf, x0, rho, beta, omega, it_max, tol):
    x = x0
    x_seq = [x0]
    for it in range(it_max):
        fx = f(x[0], x[1])
        gradfx = gradf(x[0], x[1])
        direction = -gradfx
        
        # Backtracking
        alpha = rho
        x_next = x + alpha * direction
        while f(x_next[0], x_next[1]) > fx + alpha * omega * np.dot(gradfx, direction):
            alpha *= beta
            x_next = x + alpha * direction
            
        x = x_next
        x_seq.append(x)
        if npl.norm(direction) <= tol:
            return np.array(x_seq)
    
    print("La méthode de gradient avec backtracking n'a pas convergé")
    return np.array(x_seq)
```

```python
# Exemple. 
```

```python
x0 = np.array([0., 2.])
N = 1000
tol = 1e-3
pas = 1
Xb = gradient_back(rosenbrock, grad_rosenbrock, x0, pas, 0.1, 0.1, N, tol)
print(Xb)
print(len(Xb) - 1)
```

On veut ici approcher l'argument minimum d'une fonctionnelle quadratique en dimension quelconque $J \in \mathbb{N}^*$. 
On note $\mathcal{E}_J$ la fonctionnelle définie sur $\mathbb{R}^J$ par 
\begin{equation*}
\mathcal{E}_{J}(\mathbf{u})=\dfrac{1}{2}(A_{J} \mathbf{u}, \mathbf{u}) -
(\mathbf{f}_{J}, \mathbf{u}) 
\end{equation*}
où $f_J \in \mathbb{R}^J$ est un vecteur donné et $A_{J} \in \mathcal{M}_J(\mathbb{R})$ est la matrice (symétrique) donnée par :
$$
A_J = \left( \begin{array}{cccccc}
2 & -1 &   & 0\\
-1  & \ddots & \ddots \\
& \ddots  & \ddots & -1 \\
0 & & -1 & 2
\end{array} \right),
$$
c'est-à-dire telle que $A_{i, i} = 2$, $A_{i, i+1} = A_{i, i-1} = -1$ et $A_{i,j} = 0$ si $|j - i| > 1$.

```python
# Définition de A_J en fonction de J. 
```

```python
def laplacien(J):
    return 2 * np.eye(J) - np.eye(J, k=-1) - np.eye(J, k=1)
```

## Descente de gradient à pas fixe

On propose l'implémentation suivante du gradient à pas fixe dans le cas spécifique de la minimisation d'une fonctionnelle quadratique associée à une matrice $A$ et un vecteur $b$ (ie $f(x) = \frac12 (Ax, x) - (b, x)$):

```python
def gradient_fixe(A, b, x0, rho, it_max, tol):
    """ Méthode de descente de gradient à pas fixe
    
    Pour une fonctionnelle quadratique f(x) = 1/2 <Ax, x> - <b, x>
    avec A symétrique.
    
    Paramètres
    ----------
    A: numpy.ndarray
        Matrice symétrique de taille (n, n) associée à la fonctionnelle
    b: numpy.ndarray
        Second membre de taille n associée à la fonctionnelle
    x0: numpy.ndarray
        Position initiale pour la descente, de taille n
    rho: flottant
        Le pas de descente
    it_max: entier
        Nombre maximal d'itération à effectuer
    tol: flottant
        Tolérance sur la norme de la différence de deux itérés successifs.
        Utilisée comme critère d'arrêt
        
    Retours
    -------
    x: numpy.ndarray
        Estimation de l'argument minimum de la fonctionnelle (de taille n)
    residus: numpy.ndarray
        Tableau des résidus (norme du gradient) pour chaque itération
    it: entier
        Nombre d'itérations effectuées
    """
    x = x0
    residus = []
    for it in range(it_max):
        gradient = gradEJ(A, b, x)
        x = x - rho * gradient
        residus.append(npl.norm(gradient))
        if residus[-1] <= tol:
            return x, np.array(residus), it + 1
    
    print("La méthode à pas fixe n'a pas convergé");
    return x, np.array(residus), it + 1
```

**Utiliser** cette implémentation de la méthode du gradient à pas fixe pour minimiser la fonctionnelle $\mathcal{E}_J$ et comparer son comportement avec différents pas : 1/10, 4/10, 6/10 et 1.

**Commenter :** on admettra que le rayon spectral de $A_J$ est proche de 4 lorsque $J$ est grand et on rappelera qu'un schéma itératif 
$$x_{k+1} = Mx_k + b$$
converge pour tout $x_0$ **si et seulement si** le rayon spectral de $M$ est strictement inférieur à 1.


```python
# Gradient avec pas 0.1, 0.4, 0.6 et 1.0
```

```python
import scipy.stats

J = 100
A = laplacien(J)
b = np.ones(J)
x0 = np.zeros(J)
x_exact = npl.solve(A, b)

Nmax = 100_000
tol = 1e-5

for pas in [0.1, 0.4, 0.6, 1.]:
    print("\n" + '#' * 80)
    print(f"Pour le pas {pas}:")

    x_fixe, residus_fixe, N_fixe = gradient_fixe(A, b, x0, pas, Nmax, tol)
    pente_fixe, oao_fixe, _, _, _ = scipy.stats.linregress(np.arange(N_fixe), np.log(residus_fixe))
    val_pro, vec_pro = npl.eig(np.eye(J) - pas * A)
    rayon_spectral = np.max(np.abs(val_pro))
    
    print("Nombre d'itérations pour la méthode à pas fixe : ", N_fixe)
    print('Taux de convergence mesuré : ', np.exp(pente_fixe))
    print('Rayon spectral de I - alpha*A : ', rayon_spectral)
    
    plt.figure(figsize=(10, 10))
    plt.semilogy(np.arange(N_fixe), residus_fixe, '*-')
    plt.grid()
    plt.xlabel("k")
    plt.ylabel("$res_k$")
    plt.title(f"pas = {pas}")
    plt.show()
    
```

## Descente de gradient à pas optimal 

Dans le cas spécifique de la minimisation d'une fonctionnelle $f$ quadratique associée à une matrice $A$ et un vecteur $b$ (ie $f(x) = \frac12 (Ax, x) - (b, x)$), on peut utiliser la méthode de descente de gradient à **pas optimal** qui consiste à observer que $f(x - \alpha (Ax - b))$ est quadratique en le pas $\alpha$ et donc à choisir le $\alpha$ optimal dans cette direction.

**Mettre en œuvre** la méthode du gradient à pas optimal pour minimiser la fonctionnelle $\mathcal{E}_J$.

**Comparer** (en termes de nombre d'itérations pour parvenir « à convergence »). 

```python
def gradient_opti(A, b, x0, it_max, tol):
    x = x0
    residus = []
    for it in range(it_max):
        gradient = np.dot(A, x) - b
        rho = np.dot(gradient, gradient) / np.dot(A @ gradient, gradient)
        x = x - rho * gradient
        residus.append(npl.norm(gradient))
        if residus[-1] <= tol:
            return x, np.array(residus), it + 1
    
    print("La méthode à pas optimal n'a pas convergé");
    return x, np.array(residus), it + 1
```

```python
import scipy.stats

J = 100
A = laplacien(J)
b = np.ones(J)
x0 = np.zeros(J)
x_exact = npl.solve(A, b)

Nmax = 100_000
tol = 1e-5

x_opti, residus_opti, N_opti = gradient_opti(A, b, x0, Nmax, tol)
pente_opti, oao_opti, _, _, _ = scipy.stats.linregress(np.arange(N_opti), np.log(residus_opti))

print("Nombre d'itérations pour la méthode à pas optimal : ", N_opti)
print('Taux de convergence mesuré : ', np.exp(pente_opti))

plt.figure(figsize=(10, 10))
plt.semilogy(np.arange(N_opti), residus_opti, '*-')
plt.grid()
plt.xlabel("k")
plt.ylabel("$res_k$")
plt.title("pas optimal")
plt.show()
```

## EDP elliptique

On s'intéresse au problème aux limites 
$$
\left\{ \begin{array}{l}
-u''(x) = f(x), \quad x \in ]0, 1[,\\
u(0) = u(1) = 0, 
\end{array}\right.
$$
où $f$ est une fonction donnée dans $L^2(0, 1)$. Ce problème modélise la déformation d'une corde élastique d'altitude $u(x)$ soumise à un chargement $f$. On rappelle que ce problème est bien posé, ceci peut se démontrer en appliquant le théorème de Lax-Milgram.

On peut chercher à en approcher la solution, avec une **méthode des différences finies** : on se donne $J \in \mathbb{N}^*$, on pose $h=\frac{1}{J+1}$ et $x_j=jh$ pour $j = 0,\dots,J+1$. On calcule des $u_j$ en résolvant 
$$
\left\{\begin{array}{l}
\displaystyle{}\frac{-u_{j-1}+2u_j-u_{j+1}}{h^2} = f(x_j), \quad j = 1,\dots,J, \\
u_0 = u_{J+1} = 0. 
\end{array}\right.
$$
Ce problème s'écrit sous la forme matricielle $A\mathbf{u} = F \in \mathbb{R}^J$ où $A$ est tridiagonale avec des 2 sur la diagonale et des -1 sur la sous et la sur-diagonale. 

On peut montrer que, si $u$ est une solution de classe $\mathcal{C}^4([0, 1])$ du problème aux limites, il existe un réel $C$ tel que $\sup_{j = 0,\dots,J+1}|u_j -u(x_j)| \leq C h$ pour tout $J$. 


On remarquera que la matrice $A$ est celle construite dans l'exercice 2 au facteur $\frac{1}{h^2}$ près (que l'on peut passer au second membre).

Après avoir remarqué que l'on cherche ici à annuler $A\mathbf{u} - F$, qui est le **gradient d'une fonctionnelle quadratique**,

Comparez avec la solution calculée directement en résolvant le système linéaire $Au = F$ (voir [`np.linalg.solve`](https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html)).

```python
def f(x):
    return -30 * np.ones_like(x)

L = 1. # longueur de l'intervalle en espace
J = 100 # nombre de degrés de liberté (liés aux points intérieurs)

x, h = np.linspace(0, L, J + 2, retstep=True)

A = laplacien(J)
F = h**2 * f(x[1:-1])

Nmax = 100_000
tol = 1e-6
pas = 0.4

u0 = np.zeros(J)
u_fixe, residus_fixe, N_fixe = gradient_fixe(A, F, x0, pas, Nmax, tol)
print(N_fixe, "itérations")

u_ref = npl.solve(A, F)

# On ajoute les conditions aux bords
u_fixe = np.hstack((0, u_fixe, 0))
u_ref = np.hstack((0, u_ref, 0))

# Affichage
plt.figure(figsize=(10, 10))
plt.plot(x, u_fixe, '-', label="gradient à pas fixe")
plt.plot(x, u_ref, ':', label="référence")
plt.grid()
plt.xlabel('x')
plt.ylabel('u(x)')
plt.legend()
plt.show()
```

## EDP sous contrainte

On s'intéresse à l'approximation d'un problème d'EDP sous contraintes qui modélise une corde élastique fixée en ses deux extrémités $x = 0$ et $x = 1$, soumise à un chargement et située au-dessus d'un obstacle. Lorsque la corde ne touche pas l'obstacle, son altitude satisfait à l'équation $-u'' = f$, et sinon... elle touche l'obstacle ! 

Soit $f$ et $g$ deux fonctions continues données sur
$[0,1]$. On souhaite résoudre le problème d'obstacle (donc) suivant :

Trouver $u : [0, 1] \rightarrow \mathbb{R}$ tel que 
\begin{equation}
(1)
\left\{ \begin{array}{rcl}
-u''(x) & \geq & f(x), \\
u(x) & \geq & g(x),\\
(-u''(x)-f(x))(u(x)-g(x)) & = & 0,\\
\end{array}\right\} \quad sur \,\, ]0, 1[,\,\,  et \,\,u(0) =
u(1) = 0. 
\end{equation}
 
La première équation traduit le fait que la dérivée seconde de la fonction 
$u$ est inférieure à $-f$ ; ceci est une conséquence de l'équation de la corde : $-u'' = f + R$ où $f$ est le chargement (connu) et $R$ est la réaction du support (l'obstacle), dont on ne sait qu'une chose : elle est positive ou nulle. La deuxième équation modélise la présence de l'obstacle :
on veut que la solution $u$ soit au dessus de $g$. La troisième
équation traduit le fait que l'on a au moins égalité dans une
des deux équations précédentes : 
soit on résout $-u''(x)=f(x)$ (en $x$), et la corde est libre, 
soit on résout $u(x)=g(x)$ (en $x$), et la corde est sur l'obstacle. 

On discrétise le problème par
différences finies. On introduit une subdivision uniforme $x_j=jh$
de $[0,1]$, où $h=\frac{1}{J+1}$ désigne le pas en espace du
maillage et où $i \in \left\lbrace 0,\ldots, J+1 \right\rbrace$. On
cherche alors à résoudre le problème suivant : 
\begin{equation}
(2)
\left\{ \begin{array}{rcll}
\dfrac{-u_{j-1}+2u_j-u_{j+1}}{h^2} & \geq & f(x_j), \\
u_j & \geq & g (x_j),\\
\left(\dfrac{-u_{j-1}+2u_j-u_{j+1}}{h^2}-f(x_j)\right)(u_j-g (x_j))
& = & 0,\\ 
\end{array}\right\} \;  pour \,\, j = 1 \ldots J, \,\, et \,\,
u_0 = u_{J+1} = 0. 
\label{eq:diff}
\end{equation}
  
On note $\mathcal{E}_J$ la fonctionnelle définie par 
\begin{equation*}
\mathcal{E}_{J}(\mathbf{u})=\dfrac{1}{2}(A_{J} \mathbf{u}, \mathbf{u}) -
(\mathbf{f}_{J}, \mathbf{u}) 
\end{equation*}
où $A_{J}$ et $\mathbf{f}_{J}$ sont donnés par :
$$
A_J = \left( \begin{array}{cccccc}
2 & -1 &   & 0\\
-1  & \ddots & \ddots \\
& \ddots  & \ddots & -1 \\
0 & & -1 & 2
\end{array} \right) \quad et \,\, \; \mathbf{f}_J = h^2 \left(
\begin{array}{c} 
f(x_1)\\
\vdots\\
f(x_J)
\end{array} \right).
$$

On peut alors montrer l'équivalence suivante :
$$
\mathbf{u}_J = \left( \begin{array}{c}
u_1\\
\vdots\\
u_J
\end{array} \right) \,\, est \,\, solution \,\, de \,\, (2) \,\, si \,\, et \,\,seulement \,\, si 
$$
\begin{equation}
\mathbf{u}_J \  minimise \ \ \mathcal{E}_J(\mathbf{u})
\ \  sur \ \  K_J = \{ \mathbf{v} = (v_j)_{j=1 \ldots J} : v_j \geq
g_j \; \forall \, i \}. 
\end{equation}

### Méthode de gradient projeté 
On rappelle l'algorithme du gradient projeté à pas fixe
pour minimiser une fonctionnelle $J : \mathbb{R}^J \rightarrow
\mathbb{R}$ sur un ensemble $K$, pour un point de départ
$\mathbf{u}^0$, un pas $\rho$ et un test d'arrêt $\varepsilon$
préalablement définis : 

#### Méthode du gradient projeté à pas fixe
Initialiser le résidu $r^0$ à $1$ et le compteur $k$ à $0$.

**Tant que** le résidu est plus grand que $\varepsilon$ et que le compteur n'est pas trop grand :

1. calculer la direction de descente $\mathbf{w}^k =-\nabla \mathcal{E}_J (\mathbf{u}^k)$,
2. poser $\mathbf{u}^{k+1} = P_K(\mathbf{u}^k +\rho \mathbf{w}^k)$, où $P_K$ désigne la projection orthogonale sur $K$,
3. calculer le résidu : $r^{k+1} = || \mathbf{u}^{k+1} - \mathbf{u}^k ||$,
4. incrémenter le compteur.

#### Projection
Montrer que la projection orthogonale sur $K_J$ est donnée par :
$$
P_{K_J} (\mathbf{u}) = (\max(u_j, g_j))_{j= 1 \ldots J}.
$$
Implémenter une fonction  qui prend en argument un point
$\mathbf{u}$ et $\mathbf{g}_J = (g(x_j))_{j= 1 \ldots J}$, et qui
renvoie $P_{K_J}(\mathbf{u})$. 

#### Implémentation

$A_J$,
$\mathbf{f}_J$, $\mathbf{g}_J$, un point de départ $\mathbf{u}^0$,
un pas $\rho$ et un test d'arrêt (caractérisé par un petit réel $\varepsilon$), et renverra le
vecteur $\mathbf{u}_J$ qui minimise (presque) $\mathcal{E}_J$ sur $K_J$ ainsi que le
nombre d'itérations effectuées. Un test d'arrêt classique et efficace est vérifié lorsque la norme de la différence (éventuellement divisée par $\rho$) entre deux itérés successifs est inférieure à une tolérance donnée.

### II) Méthode de pénalisation (pour aller plus loin !)

La méthode de pénalisation consiste à remplacer le
problème de minimisation sous contrainte (1)
par une suite de problèmes de minimisation sans contrainte qui 
« converge » vers (1).
Soit $\eta>0$ donné, et $\mathcal{E}_J^\eta$ défini par :
$$
\mathcal{E}_J^\eta (\mathbf{u}) = \mathcal{E}_J (\mathbf{u}) + \frac{1}{\eta}
\sum_{j=1}^J (\max(g_j - u_j, 0))^2 .
$$ 
Le minimiseur $\mathbf{u}_J^\eta$ de $\mathcal{E}_J^\eta$ sur $\mathbb{R}^J$
converge vers $\mathbf{u}_J$ quand $\eta$ tend vers $0$.

On admet que le gradient de $\mathcal{E}_J^\eta$ est donné par :
$$
\nabla \mathcal{E}_J^\eta (\mathbf{u}) = A_J \mathbf{u} - \mathbf{f}_J -
\left( \frac{2}{\eta} \operatorname{signe}(g_j - u_j) \max (g_j - u_j, 0)
\right)_{j=1 \ldots J}
$$ 
où la fonction signe est définie par : $\operatorname{signe} (x) =
\left\{ \begin{array}{cl} 
-1 &  si \,\, x < 0,\\
0 &  si \,\, x = 0,\\
1 &  si \,\, x > 0.
\end{array} \right.$

#### Implémentation
Adapter le programme de gradient à pas fixe pour écrire

point de départ $\mathbf{u}^0$, un pas $\rho$ et un test d'arrêt
$\varepsilon$, et qui renvoie le vecteur $\mathbf{u}_J$ qui minimise 
(presque) $\mathcal{E}_J^\eta$ sur $\mathbb{R}^J$ ainsi que le nombre 
d'itérations effectuées.
